```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms
```

```python
class data_toy(Dataset):

    def __init__(self,length=100,n_features=2,transform=None):

        self.x = 2 * torch.ones(length,n_features)
        self.y = torch.ones(length,1)
        self.transform = transform
        self.len = length

    def __getitem__(self, index):
        sample = self.x[index], self.y[index]
        if self.transform:
            sample = self.transform(sample)
        return sample
    
    def __len__(self):
        return self.len
```

```python
class mult(object):
    def __init__(self,mult=2):
        self.mult = mult
    def __call__(self, sample):
        x = sample[0]
        y = sample[1]
        x = x * self.mult
        y = y * self.mult
        return x,y
```

```python
class add_mult(object):
    def __init__(self,mult=4,add=3):
        self.add = add
        self.mult = mult
    def __call__(self,sample):
        x = sample[0]
        y = sample[1]
        x = x + self.add
        y = y * self.mult
        return x,y
```

```python
data_transform = transforms.Compose([mult(),add_mult()])
```

```python
dataset_ = data_toy(length=10,transform=data_transform)
```

```python
for i in range(10):
    x,y = dataset_[i]
    print(f"sample {i} \n\tX : {x}\n\ty : {y}")
```
