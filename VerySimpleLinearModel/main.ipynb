```python
import torch
import torch.nn as nn
```

```python
class linear_network(nn.Module):

    def __init__(self, in_features,out_features):
        super(linear_network,self).__init__()
        self.linear = nn.Linear(in_features,out_features)

    def forward(self,x):
        return self.linear(x)
```

```python
x = torch.tensor([[1.,2.,3.,4.,5.],[3.,4.,5.,6.,7.]])
```

```python
model = linear_network(5,1)
y_hat = model(x)
y_hat
```

```python
model.state_dict()['linear.weight'].data[0,0] = torch.tensor([5.])
model.state_dict()['linear.weight'].data[0,1] = torch.tensor([4.])
model.state_dict()['linear.weight'].data[0,2] = torch.tensor([9.])
model.state_dict()['linear.weight'].data[0,3] = torch.tensor([8.])
model.state_dict()['linear.weight'].data[0,4] = torch.tensor([2.])
model.state_dict()['linear.bias'].data[0] = torch.tensor([2.])
```

```python
y_hat = model(x)
y_hat
```
