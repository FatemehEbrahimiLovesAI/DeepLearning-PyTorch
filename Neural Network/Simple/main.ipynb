```python
import torch
from torch.utils.data import DataLoader, Dataset
from torch import nn,optim,sigmoid
```

```python
class Net(nn.Module):
    def __init__(self, D_in , H , D_out): # H : تعداد نورون ها
        super(Net, self).__init__()
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H , D_out)
    def forward(self,X):
        X = sigmoid(self.linear1(X))
        X =sigmoid(self.linear2(X))
        return X
```

```python
model = Net(1,2,1)
```

```python
class Data(Dataset):
    def __init__(self,train=True) -> None:
        self.X = torch.randn(1000).view(-1,1)
        self.f = X * 2 + -3
        self.y = self.f + 0.1 * torch.randn(self.X.size())
        if train == False:
            self.y = self.f + 0.1 * torch.randn(self.X.size())

    def __getitem__(self, index):
        return self.X[index], self.y[index]
    def __len__(self):
        return len(self.X)
```

```python
train_data = Data()
validate_data = Data(train=False)
```

```python
train_loader = DataLoader(train_data,batch_size=100)
validate_loader = DataLoader(validate_data,batch_size=100)
```

```python
optimizer = optim.SGD(model.parameters(),lr=0.1,momentum=0.9)
```

```python
epoches = 100
hist_train_loss , hist_validate_loss = [] , []
```

```python
for epoch in range(epoches):
    mean_train_loss , mean_validate_loss = 0,0

    # ================== Train Model =======================
    model.train()
    for X_batch , y_batch in train_loader:

        y_hat = model(X_batch)
        loss = nn.functional.mse_loss(y_hat,y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        mean_train_loss += loss.item()

    mean_train_loss = mean_train_loss / len(train_data)
    hist_train_loss.append(mean_train_loss)

    # ================= Validate Model ====================

    model.eval()
    with torch.no_grad():
        for X_batch, y_batch in validate_loader:

            y_hat = model(X_batch)
            loss = nn.functional.mse_loss(y_hat,y_batch)
            mean_validate_loss += loss.item()

        mean_validate_loss = mean_validate_loss / len(validate_data)
        hist_validate_loss.append(mean_validate_loss)
```

```python
print(f' Train Loss : {torch.mean(torch.tensor(hist_train_loss))}\n  Validation Loss : {torch.mean(torch.tensor(hist_validate_loss))}')
```
