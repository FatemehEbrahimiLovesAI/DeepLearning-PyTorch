```python
import torch
from torch.utils.data import DataLoader,Dataset
from torch import nn,optim
```

```python
class Data(Dataset):
    def __init__(self) -> None:

        self.X = torch.linspace(-20,20,100).view(-1,1)
        self.y = torch.zeros(self.X.shape[0])
        self.y[(self.X[:,0] > -10) & (self.X[:,0] < -5)] = 1
        self.y[(self.X[:,0] > 5) & (self.X[:,0] < 10)] = 1
        self.y = self.y.view(-1,1)

    def __getitem__(self, index):
        return self.X[index], self.y[index]

    def __len__(self):
        return self.X.shape[0]
```

```python
class Net(nn.Module):
    def __init__(self, d_in, h, d_out) -> None:
        super(Net, self).__init__()
        self.linear1 = nn.Linear(d_in,h)
        self.linear2 = nn.Linear(h,d_out)

    def forward(self,X):
        X = nn.functional.sigmoid(self.linear1(X))
        X = nn.functional.sigmoid(self.linear2(X))
        return X
```

```python
loss_fn = nn.BCELoss()
train_set = Data()
train_loader = DataLoader(train_set,batch_size=100)
```

```python
model = Net(1,12,1)
optimizer = optim.Adam(model.parameters(),lr=0.01)
```

```python
def train_model(model, train_loader,epoches,optimizer, loss_fn):
    cost = []


    for epoch in range(epoches):
        total = 0
        for X_batch , y_batch in train_loader:
            optimizer.zero_grad()
            y_hat = model(X_batch)
            loss = loss_fn(y_hat , y_batch)
            loss.backward()
            optimizer.step()
            total += loss.item()
        cost.append(total)
    return cost
```

```python
cost = train_model(model,train_loader,5,optimizer,loss_fn)
cost
```
