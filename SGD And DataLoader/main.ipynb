# SGD : Stochastic Gradient Decent

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,Dataset
```

```python
X = torch.arange(-3.,3.,0.1).view(-1,1)

w = torch.tensor(-1.,requires_grad=True)
b = torch.tensor(3.,requires_grad=True)

Y = 4 * X + 5
```

```python
def forward(x):
    return w * x + b
```

```python
def mse(y,y_hat):
    return torch.mean((y_hat-y)**2)
```

## Whithout DataLoader

```python
lr = 0.1
cost = []

for epoch in range(4):
    for x,y in zip(X,Y):
        y_hat = forward(x)
        loss = mse(y,y_hat)
        cost.append(loss.data)

        loss.backward()
        w.data = w.data - lr * w.grad.data
        b.data = b.data - lr * b.grad.data

        w.grad.data.zero_()
        b.grad.data.zero_()
print(torch.mean(torch.tensor(cost)))
```

## With DataLoader

```python
class Data(Dataset):
    def __init__(self):
        self.x = torch.arange(-3.,3.,0.1).view(-1,1)
        self.y = 4 * self.x + 5


    def __getitem__(self, index):
        return self.x[index] , self.y[index]
    
    def __len__(self):
        return len(self.x)
```

```python
dataset = Data()
train_dataloader = DataLoader(dataset=dataset,batch_size=1)
```

```python
w = torch.tensor(-1.,requires_grad=True)
b = torch.tensor(3.,requires_grad=True)
```

```python
lr = 0.1
cost = []

for epoch in range(4):
    for x,y in train_dataloader:
        y_hat = forward(x)
        loss = mse(y,y_hat)
        cost.append(loss.data)

        loss.backward()
        w.data = w.data - lr * w.grad.data
        b.data = b.data - lr * b.grad.data

        w.grad.data.zero_()
        b.grad.data.zero_()

print(torch.mean(torch.tensor(cost)))
```
