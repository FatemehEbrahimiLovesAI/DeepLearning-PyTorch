# Import Liberaries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import TensorDataset,DataLoader
import torch.nn as nn
from torch.optim import SGD
```

# Data Handeling

## Load Dataset

```python
dataset = pd.read_csv(r'./Salary Data.csv')
```

## features and label Split

```python
X,y = dataset['Experience Years'].values, dataset['Salary'].values
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=12)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Convert Vector to Matrix

```python
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)

X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Preprosessing

```python
X_scaler = StandardScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)

y_scaler = StandardScaler()
y_train = y_scaler.fit_transform(y_train)
y_test = y_scaler.transform(y_test)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Convert Numpy Array to Tensors

```python
X_train = torch.tensor(X_train,dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)

y_train = torch.tensor(y_train,dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Create Tensor Dataset

```python
train_set = TensorDataset(X_train,y_train)
test_set = TensorDataset(X_test,y_test)
```

## Create Data Loader

```python
train_loader = DataLoader(train_set,batch_size=5,shuffle=True)
test_loader = DataLoader(test_set,batch_size=4)
```

# Train And Test

## Train

### Create Model

```python
linear_regression = nn.Linear(1,1)
```

### Loss 

```python
mse = nn.MSELoss()
```

### Optimizer

```python
optimizer = SGD(linear_regression.parameters(),lr=0.001,momentum=0.9)
```

### Training

```python
n_epoches = 100

for epoch in range(n_epoches):
    for X_batch,y_batch in train_loader:
        # model predict
        y_hat = linear_regression(X_batch)
        # loss
        loss = mse(y_hat,y_batch)
        # gradient and optimizer
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    # Print output
    print(f"Epoch {epoch+1}\n\tWeights : '{linear_regression.weight.item()}' \n\tBias : '{linear_regression.bias.item()}' \n\tLoss : '{loss.item()}' \n\tWeights Grads : '{linear_regression.weight.grad.item()}' \n\tBias Grads '{linear_regression.bias.grad.item()}'")
```

### Save model

```python
torch.save(linear_regression,f='./LinearRegressionModel')
```

## Test

### load model

```python
model = torch.load(f='./LinearRegressionModel', weights_only=False)
```

### Get loss

```python
for X_batch,y_batch in test_loader:
    y_hat = model(X_batch)
    loss = nn.functional.mse_loss(y_hat,y_batch)
    print(loss)
```
